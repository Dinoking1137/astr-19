Create a Jupyter notebook, import matplotlib. Write cells that 1) create a numpy array x ranging from [0,1] in 100 steps and 2) that defines functions that return sin(x) and cos(x).  In a new cell use matplotlib to create a multipanel plot (1 row, 2 columns), plotting sin(x) vs. x in the left panel and cos(x) vs. x in the right panel. Label the panels with sin(x) and cos(x), and save the figure as a PDF.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 1, 100) #creates the numpy array from [0, 1] inclusive in 100 steps

def sinx(x): #returns the sine of whatever values are inputted
    return np.sin(x)
def cosx(x): #returns the cosine of whatever values are inputted
    return np.cos(x)
```

```python
# call subplots to generate a multipanel figure
# this means 1 row, 2 columns figure
figure, axes = plt.subplots(1, 2)

# figure is the entire space used by the figure
# axes is a list where axes[0] and axes[1] are the subplots

# first panel
axes[0].plot(sinx(x), x)
axes[0].set_xlabel('sin(x)')
axes[0].set_ylabel('x')
axes[0].set_title('sin(x)')
                   
# second panel
axes[1].plot(cosx(x), x)
axes[1].set_xlabel('cos(x)')
axes[1].set_ylabel('x')
axes[1].set_title('cos(x)')

#add more space between the subplots
figure.subplots_adjust(wspace=0.8)

plt.savefig('code_journal_2-2.pdf', bbox_inches='tight', dpi = 200)

plt.show()
```
